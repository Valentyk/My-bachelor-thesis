```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def geodetic_normal(h1,h2,h3):
    h = (h1**2 + h2**2 + h3**2)**(1/2)
    x = (h1*np.sin(h))/h
    y = (h2*np.sin(h))/h
    z = (h3*np.sin(h))/h
    w = np.cos(h)
    return(x, y, z, w)

def spherical_1(phi, theta, zeta):
    x = np.sin(phi)
    y = np.cos(phi)*np.sin(theta)
    z = np.cos(phi)*np.cos(theta)*np.sin(zeta)
    w = np.cos(phi)*np.cos(theta)*np.cos(zeta)
    return(x, y, z, w)
```

```python
geodetic_normal(2*np.pi, 1, 0)
```

```python
x = np.linspace(0, np.pi, 10000)
delta = 0

for i in range (len(geodetic_normal(0,0,0))):
    plt.plot(x, geodetic_normal(x/np.sqrt(2), x/np.sqrt(2), 0)[i], label = f"{i}")
    delta += geodetic_normal(x, 0, 0)[i]**2
delta**(1/2)
plt.plot(x, delta)
plt.legend()
```

```python
x = np.linspace(-np.pi/2, np.pi/2, 10000)

for i in range (len(geodetic_normal(0,0,0))):
    plt.plot(x, spherical_1(x, 0, 0)[i], label = f"{i}")
plt.legend()
```
